# Major Pairwise Comparison

```python
# t-test / wilcoxon
for i, k in enumerate(df.columns):
  if (i < 3): continue
  print("///", k ,"///")
  print('Shapiro-Wilk Normality Test')
  stat1, p1 = scipy.stats.shapiro(df[df['method'] == 'Holistic'][k])
  print('Statistics=%.4f, p=%.4f' % (stat1, p1))
  stat2, p2 = scipy.stats.shapiro(df[df['method'].isna()][k])
  print('Statistics=%.4f, p=%.4f' % (stat2, p2))

  print('Levene Test for Equal Variance')
  stat_eqv, p_eqv = scipy.stats.levene(df[df['method'] == 'Holistic'][k], df[df['method'].isna()][k])
  print('Statistics=%.4f, p=%.4f' % (stat_eqv, p_eqv))
  
  # if (p1 > 0.05 and p2 > 0.05 and p_eqv > 0.05):
  if (p1 > 0.05 and p2 > 0.05):
    print('\n⫸ Parametric')
  else:
    print('\n⫷ Non-parametric')
  # print('Both samples are normally distributed')
  print('\n# Paired t-test')
  stat, p= scipy.stats.ttest_rel(
                              a = df[df['method'] == 'Holistic'][k], 
                              b = df[df['method'].isna()][k],
                              alternative='greater'
                            )
  print("\nt-test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Holistic has greater', k)
  else:
    print('Holistic has no greater', k)
  print('-◎-')
  stat, p= scipy.stats.ttest_rel(
                              a = df[df['method'].isna()][k],
                              b = df[df['method'] == 'Holistic'][k], 
                              alternative='greater'
                            )
  print("t-test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Baseline has greater', k)
  else:
    print('Baseline has no greater', k)
  print("\n--◎--◎--◎--")
  stat, p= scipy.stats.ttest_rel(
                              a = df[df['method'] == 'Holistic'][k], 
                              b = df[df['method'].isna()][k],
                              alternative='less'
                            )
  print("\nt-test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Holistic has less', k)
  else:
    print('Holistic has no less', k)
  print('-◎-')
  stat, p= scipy.stats.ttest_rel(
                              a = df[df['method'].isna()][k],
                              b = df[df['method'] == 'Holistic'][k], 
                              alternative='less'
                            )
  print("t-test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Baseline has less', k)
  else:
    print('Baseline has no less', k)
  print("\n--◎--◎--◎--")
  stat, p= scipy.stats.ttest_rel(
                              a = df[df['method'] == 'Holistic'][k], 
                              b = df[df['method'].isna()][k],
                              alternative='two-sided'
                            )
  print("\nt-test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Holistic and Baseline are different in', k)
  else:
    print('Holistic and Baseline have no different in', k)
  # print("\n# Wilcoxon")
  # stat, p= scipy.stats.wilcoxon(
  #                             x = df[df['method'] == 'Holistic'][k], 
  #                             y = df[df['method'].isna()][k], 
  #                             alternative='greater'
  #                           )
  # print("Wilcoxon signed rank sum test: ", stat, " | p-value", p)
  # if (p < 0.05):
  #   print('✅Holistic has greater', k)
  # else:
  #   print('Holistic has not greater', k)
  # print('--')
  # stat, p= scipy.stats.wilcoxon(
  #                             x = df[df['method'].isna()][k], 
  #                             y = df[df['method'] == 'Holistic'][k], 
  #                             alternative='greater'
  #                           )
  # print("Wilcoxon signed rank sum test: ", stat, " | p-value", p)
  # if (p < 0.05):
  #   print('✅Holistic has less', k)
  # else:
  #   print('Holistic has not less', k)
  # print("\n# Sign Test for ordinal data")
  # # sign test
  # n = len(df[df['method'] == 'Holistic'][k])
  # x = sum(1 for idx, ele in enumerate(df[df['method'] == 'Holistic'][k].to_numpy()) if ele > df[df['method'].isna()][k].to_numpy()[idx])
  # p = scipy.stats.binom_test(x, n, alternative='greater')
  # if (p < 0.05):
  #   print('✅Holistic has greater', k)
  # else:
  #   print('Holistic has not greater', k)
  # print("--")
  # n = len(df[df['method'].isna()][k])
  # x = sum(1 for idx, ele in enumerate(df[df['method'].isna()][k].to_numpy()) if ele > df[df['method'] == 'Holistic'][k].to_numpy()[idx])
  # p = scipy.stats.binom_test(x, n, alternative='greater')
  # if (p < 0.05):
  #   print('✅Holistic has less', k)
  # else:
  #   print('Holistic has not less', k)
# else:
  # print('Not all samples are normally distributed')
  print('\n# Wilcoxon')
  stat, p= scipy.stats.wilcoxon(
                              x = df[df['method'] == 'Holistic'][k], 
                              y = df[df['method'].isna()][k], 
                              alternative='greater'
                            )
  print("\nWilcoxon signed rank sum test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Holistic has greater', k)
  else:
    print('Holistic has no greater', k)
  print('-◎-')
  stat, p= scipy.stats.wilcoxon(
                              x = df[df['method'].isna()][k], 
                              y = df[df['method'] == 'Holistic'][k], 
                              alternative='greater'
                            )
  print("Wilcoxon signed rank sum test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Baseline has greater', k)
  else:
    print('Baseline has no greater', k)
  print("\n--◎--◎--◎--")
  stat, p= scipy.stats.wilcoxon(
                              x = df[df['method'] == 'Holistic'][k], 
                              y = df[df['method'].isna()][k], 
                              alternative='less'
                            )
  print("\nWilcoxon signed rank sum test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Holistic has less', k)
  else:
    print('Holistic has no less', k)
  print('-◎-')
  stat, p= scipy.stats.wilcoxon(
                              x = df[df['method'].isna()][k], 
                              y = df[df['method'] == 'Holistic'][k], 
                              alternative='less'
                            )
  print("Wilcoxon signed rank sum test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Baseline has less', k)
  else:
    print('Baseline has no less', k)
  print("\n--◎--◎--◎--")
  stat, p= scipy.stats.wilcoxon(
                              x = df[df['method'] == 'Holistic'][k], 
                              y = df[df['method'].isna()][k], 
                              alternative='two-sided'
                            )
  print("\nWilcoxon signed rank sum test: ", stat, " | p-value", p)
  if (p < 0.05):
    print('✅Holistic and Baseline are different in', k)
  else:
    print('Holistic and Baseline have no different in', k)
  print("\n\n")
```

# Basic Info

```python
# some basic information
for i, k in enumerate(df.columns):
  if (i < 3): continue
  print("///", k ,"///")
  mh = df[df['method'] == 'Holistic'][k].mean()
  mn = df[df['method'].isna()][k].mean()
  print("Holistic: ", mh)
  print("None: ", mn)
  
  if (mh > mn):
    print('🌐Holistic has greater mean in', k)
    print('Holistic is', mh - mn, 'greater than None')
    print('Holistic is', round(100*(mh-mn)/mn,2), 'percent greater than None')
  else:
    print('None has greater mean in', k)
    print('None is', mn - mh, 'greater than Holistic')
    print('None is', round(100*(mn-mh)/mh,2), 'percent greater than Holistic')

  print("\n\n")
```
